```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is the DataFrame, and 'Churn Label' indicates if a customer churned (Yes/No)
# Calculate the count of churns grouped by 'Senior Citizen' and 'Churn Label'
churn_counts = df.groupby(['Senior Citizen', 'Churn Label']).size().unstack()

# Calculate the percentage
churn_percent = churn_counts.divide(churn_counts.sum(axis=1), axis=0) * 100

# Plotting the stacked bar chart
ax = churn_percent.plot(kind='bar', stacked=True, figsize=(5, 5), colormap='viridis')

# Adding percentage labels to the bars
for p in ax.patches:
    width = p.get_width()    # Get the width of the bar
    height = p.get_height()  # Get the height of the bar (percentage value)
    x = p.get_x()            # Get the x-coordinate of the bar
    y = p.get_y() + height / 2  # Get the y-coordinate of the bar's center

    # Annotate the bar with the percentage value if height is not zero
    if height > 0:
        ax.text(x + width / 2, y, f'{height:.1f}%', ha='center', va='center', fontsize=9, color='black')

# Adding labels and title
plt.title("Churn by Senior Citizen (Percentage)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Assuming your DataFrame is named 'df'
# List of columns for which you want to create countplots
columns = [
    'Phone Service', 'Multiple Lines', 'Internet Service', 'Online Security',
    'Online Backup', 'Device Protection', 'Tech Support', 'Streaming TV',
    'Streaming Movies'
]

# Number of rows and columns for the subplots layout
n_rows = 3
n_cols = 3

# Create a figure and axes for the subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 12))
fig.suptitle('Countplots for Various Services', fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot countplots for each column
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue="Churn Label")
    axes[i].set_title(f'Countplot of {col}')

# Hide any unused subplots if the number of columns is less than the number of subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust the layout
plt.tight_layout()
plt.subplots_adjust(top=0.93)  # Adjust top to fit the suptitle
```

The charts display churn analysis for various telecom services, with the counts split by "Yes" or "No" for customer churn. Each subplot represents different services, such as phone service, internet service, online security, and streaming options. Generally, customers who did not use certain services (like tech support or online security) exhibited higher churn rates. Services like fiber optic internet also showed higher churn compared to DSL or no internet services.
